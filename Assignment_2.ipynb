Import necessary libraries

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

Load the car_crashes dataset from Seaborn

```python
car_crashes = sns.load_dataset('car_crashes')
```

**Data Vizuilation**

1) Pair Plot

```python
sns.pairplot(car_crashes)
plt.title('pairplot of car crashes dataset')
plt.show()
```

*Pairplot helps us visualize the relationships between different numerical variables. We can observe trends, correlations, and potential outliers.*

2) Scatterplot of alcohol vs total



```python
sns.scatterplot(x='alcohol',y='total', data=car_crashes)
plt.title('scatterplot of alochol vs total crashes ')
plt.xlabel('alcohol consumption')
plt.ylabel('total crashes')
plt.show()
```

*This scatterplot shows a possible positive correlation between alcohol consumption and total crashes.*

3) Boxplot of speeding

```python
sns.boxplot(x='speeding' , y ='total' , data=car_crashes)
plt.title('Boxplot of Speeding vs. Total Crashes')
plt.xlabel('Speeding (Percentage)')
plt.ylabel('Total Crashes')
plt.show()
```

*The boxplot helps us understand the distribution of total crashes among different speeding percentages.*


4) Histogram of 'total'

```python
plt.hist(car_crashes['total'], bins=15, color='skyblue', edgecolor='black')
plt.title('Histogram of Total Crashes')
plt.xlabel('Total Crashes')
plt.ylabel('Frequency')
plt.show()
```

*The histogram provides an overview of the distribution of total crashes.
*

5) Barplot of 'abbrev' vs. 'total'

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='abbrev', y='total', data=car_crashes, palette='viridis')
plt.title('Barplot of Total Crashes by State')
plt.xlabel('State Abbreviation')
plt.ylabel('Total Crashes')
plt.xticks(rotation=90)
plt.show()
```

*This barplot shows the total crashes for each state, allowing us to identify states with higher and lower crash rates.
*

6)  Violinplot of 'alcohol' by 'ins_premium

```python
plt.figure(figsize=(10, 6))
sns.violinplot(x='ins_premium', y='alcohol', data=car_crashes, palette='coolwarm')
plt.title('Violinplot of Alcohol Consumption by Insurance Premium')
plt.xlabel('Insurance Premium')
plt.ylabel('Alcohol Consumption')
plt.show()
```

*The violinplot helps us visualize the distribution of alcohol consumption based on different insurance premium levels.
*

7) Heatmap of correlation matrix

```python
correlation_matrix = car_crashes.corr()
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

*The heatmap shows the correlation between numerical variables. We can see that alcohol consumption is positively correlated with total crashes.*

```python
# Conclusion
# Based on the visualizations and analysis:
# - There is a positive correlation between alcohol consumption and total crashes.
# - Some states have notably higher total crashes than others, as seen in the barplot.
# - The violinplot suggests that there may not be a strong relationship between alcohol consumption and insurance premiums.
# - The correlation heatmap confirms the positive correlation between alcohol and total crashes.

# These insights can be useful for policymakers and insurance companies to understand the factors affecting car crashes.
```
